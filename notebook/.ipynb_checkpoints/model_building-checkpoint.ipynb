```python
df.fillna(0,inplace=True)
```

```python
df_copy=df.copy()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
df_copy['BRANCH']=encoder.fit_transform(df_copy['BRANCH'])
df_copy['CATEGORIES']=encoder.fit_transform(df_copy['CATEGORIES'])
```

```python
x=df_copy.drop(columns=['COLLEGE','Cutoff_2024'])
y=df_copy['Cutoff_2024']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rm=RandomForestRegressor()
rm.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(rm.predict(x_test),y_test)
```
